```python
import numpy
datadir = '../data/'
```

```python
def get_time_mem(file_name):
    file_obj = open(file_name,'r')
    for line in file_obj:
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            time = fields1[0]
            fields2 = line.split('avgdata ')
            fields3 = fields2[1].split('max')
            mem = fields3[0]
            return float(time),float(mem)
```

```python
# read time files

mkindex_time={}
mkindex_mem={}
suffix_time={}
suffix_mem={}

for i1 in range(1,4):
    for i2 in [1,4]:
        for i3 in [22,31]:
            name = str(i1)+'_'+str(i2)+'_'+str(i3)
            filename = datadir +'mkindex_'+str(i1)+'_part'+str(i2)+'_'+str(i3)+'.time'
            mkindex_time[name],mkindex_mem[name] = get_time_mem(filename)

        name = str(i1) + '_' + str(i2)
        filename = datadir +'suffix_'+str(i1)+'_part'+str(i2)+'.time'
        suffix_time[name],suffix_mem[name] = get_time_mem(filename)

        
for i1 in [4,5]:
    for i3 in [22,31]:
        name = str(i1)+'_4_'+str(i3)
        filename = datadir +'mkindex_'+str(i1)+'_part4_'+str(i3)+'.time'
        mkindex_time[name],mkindex_mem[name] = get_time_mem(filename)
    name = str(i1) + '_4'
    filename = datadir +'suffix_'+str(i1)+'_part4.time'
    suffix_time[name],suffix_mem[name] = get_time_mem(filename)


jellyfish_mem = {}
jellyfish_time = {}
for i1 in range(1,6):
    for i2 in [22,31]:
        name = str(i1) + '_' + str(i2)
        filename = datadir +str(i1)+'_'+str(i2)+'_jellyfish.time'
        jellyfish_time[name],jellyfish_mem[name] = get_time_mem(filename)

dsk_mem = {}
dsk_time = {}
for i1 in range(1,6):
    for i2 in [22,31]:
        name = str(i1) + '_' + str(i2)
        filename = datadir+'dsk_'+str(i1)+'_'+str(i2)+'.time'
        dsk_time[name], dsk_mem[name] = get_time_mem(filename)
        

khmer_mem = {}
khmer_time = {}

for i1 in range(1,6):
    for i2 in [0.01,0.05]:
        for i3 in [22,31]:
            name = str(i1)+'_'+str(i2)+'_'+str(i3)
            filename = datadir +'bloom.'+str(i1)+'_'+str(i2)+'_'+str(i3)+'.time'
            khmer_time[name],khmer_mem[name] = get_time_mem(filename)
```

Figure 5
=======
Percentage of the unique k-mers starting in different position in reads

```python

file17 = open(datadir+"ecoli_ref.fastq.fasta.pos.abund=1",'r')
file32 = open(datadir+"ecoli_ref.fastq.fasta.pos32.abund=1",'r')

list1 = []
list2 = []
x1 = []
x2 = []

for line in file17:
    line = line.rstrip()
    fields = line.split()
    if fields[1] != '0':
        x1.append(float(fields[0]))
        list1.append(float(fields[1]))

for line in file32:
    line = line.rstrip()
    fields = line.split()
    if fields[1] != '0':
        x2.append(float(fields[0]))
        list2.append(float(fields[1]))

plt.xlabel("starting position of k-mer")
plt.ylabel("number of abund=1 k-mers at that position")
plt.title("low-abundance k-mer distribution within reads")


plt.plot(x1,list1,'ro-')
plt.plot(x2,list2,'bo-')
plt.legend(('k=17','k=32'),)
plt.ylim(0,10000000)
plt.xlim(0,100)

plt.savefig("../figure/perc_unique_pos.pdf")
```
