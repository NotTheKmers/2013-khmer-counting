```python
import numpy
datadir = '../data/'
```

```python
def get_time_mem(filename):
    "Extract the user time and max memory as generated by 'time' command"
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            time = fields1[0]
            fields2 = line.split('avgdata ')
            fields3 = fields2[1].split('max')
            mem = fields3[0]
            return float(time), float(mem)
```

```python
# read time files

mkindex_time={}
mkindex_mem={}
suffix_time={}
suffix_mem={}

# tallymer runtime info
for i1 in range(1,6):
    for i2 in [1,4]:
        for i3 in [22,31]:
            name = "%d_%d_%d" % (i1, i2, i3)
            filename = 'mkindex_%d_part%d_%d.time' % (i1, i2, i3)
            filename = datadir + filename
            #print filename
            mkindex_time[name],mkindex_mem[name] = get_time_mem(filename)

        name = "%d_%d" % (i1, i2)
        filename = datadir + 'suffix_%d_part%d.time' % (i1, i2)
        #print filename
        suffix_time[name],suffix_mem[name] = get_time_mem(filename)
#print "Tallymer suffix"
#for key in sort(suffix_time.keys()):
#    print key, suffix_time[key],suffix_mem[key]
#print "Tallymer mkindex"
#for key in sort(mkindex_time.keys()):
#    print key, mkindex_time[key],mkindex_mem[key]
    
#print mkindex_time
#print mkindex_mem
#print 'hello'

jelly_count_mem = {}
jelly_count_time = {}
for i1 in range(1,6):
    for i2 in [22,31]:
        name = "%d_%d" % (i1, i2)
        filename = datadir + 'jelly_%d_%d.time1' % (i1, i2)
        jelly_count_time[name],jelly_count_mem[name] = get_time_mem(filename)

#print jelly_count_mem
#print jelly_count_time
#print "Jellyfish count"
#for key in sort(jelly_count_mem.keys()):
#    print key, jelly_count_time[key],jelly_count_mem[key]
    
    
jelly_hist_mem = {}
jelly_hist_time = {}
for i1 in range(1,6):
    for i2 in [22,31]:
        name = "%d_%d" % (i1, i2)
        filename = datadir + 'jelly_%d_%d.time2' % (i1, i2)
        jelly_hist_time[name],jelly_hist_mem[name] = get_time_mem(filename)
#print jelly_hist_mem
#print jelly_hist_time
#print "Jellyfish hist"
#for key in sort(jelly_hist_mem.keys()):
#    print key, jelly_hist_time[key],jelly_hist_mem[key]
    
        
dsk_mem = {}
dsk_time = {}
for i1 in range(1,6):
    for i2 in [22,31]:
        name = "%d_%d" % (i1, i2)
        filename = datadir+'dsk_%d_%d.time' % (i1, i2)
        dsk_time[name], dsk_mem[name] = get_time_mem(filename)
#print dsk_mem
#print dsk_time

#print "Dsk"
#for key in sort(dsk_mem.keys()):
#    print key, dsk_time[key],dsk_mem[key]
    

khmer_mem1 = {}
khmer_time1 = {}

for i1 in range(1,6):
    for i2 in [1,5]:
        for i3 in [22,31]:
            name = "%d_%d_%d" % (i1, i2, i3)
            filename = datadir +'bloom_%d_%d_%d.time1' % (i1, i2, i3)
            khmer_time1[name],khmer_mem1[name] = get_time_mem(filename)
#print "khmer_1"
#for key in sort(khmer_time1.keys()):
#    print key, khmer_time1[key],khmer_mem1[key]
    
    
khmer_mem2 = {}
khmer_time2 = {}

for i1 in range(1,6):
    for i2 in [1,5]:
        for i3 in [22,31]:
            name = "%d_%d_%d" % (i1, i2, i3)
            filename = datadir +'bloom_%d_%d_%d.time2' % (i1, i2, i3)
            khmer_time2[name],khmer_mem2[name] = get_time_mem(filename)
#print "khmer_2"
#for key in sort(khmer_time2.keys()):
#    print key, khmer_time2[key],khmer_mem2[key]
    
    
# other print table, out for other stuffs , draw graphs. 
            
#khmer_mem1, khmer_time1,khmer_mem2, khmer_time2
```

```python
# number of unique 22-mers in different data sets
# 
def get_total_kmers(filename):
    "Extract the user time and max memory as generated by 'time' command"
    total = 0
    for line in open(filename):
        line = line.rstrip()
        fields = line.split()
        total = total + int(fields[1])
    return total
total_list = []
for i in range(1,6):
    filename = datadir+'jelly_%d_22.hist' % (i)
    total = get_total_kmers(filename)
    total_list.append(total)

print total_list
```

```python
# 1. user time
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)

L_khmer = [] # 1% error rate , k=22
L_tallymer = [] # k=22, pick biggest memory, suffix 1 part and 4 parts different memory ,but all smaller than mkindex step. 
L_jellyfish_k22 = [] # use k=22, but memory change with different k size. k=31. 
L_dsk = [] # use k=22.

for i in range(1,6):
    L_khmer.append(khmer_time1[str(i)+'_1_22']+khmer_time2[str(i)+'_1_22'])
    L_tallymer.append(suffix_time[str(i)+'_1']+mkindex_time[str(i)+'_1_22'])
    L_jellyfish_k22.append(jelly_count_time[str(i)+'_22'] + jelly_hist_time[str(i)+'_22'])
    L_dsk.append(dsk_time[str(i)+'_22'])



ax.set_xlabel('total number of distinct k-mers')
ax.set_ylabel('Time Usage(s)')


ax.plot(total_list,L_khmer,'bo-', label='khmer')
ax.plot(total_list,L_tallymer,'go-', label='Tallymer')
ax.plot(total_list,L_jellyfish_k22,'ro-', label='Jellyfish')
ax.plot(total_list,L_dsk,'yo-', label='DSK')

ax.set_xlim(0,3e9)
ax.set_ylim(0,6000)
ax.legend(loc='best')
fig_file = '../figure/time_benchmark.pdf'
plt.savefig(fig_file)
```

Figure 2
=======
Memory usage of different k-mer counting tools

```python

# 2. memory usage of different k-mer counting tools

fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)

L_khmer_1p = [] # 1% error rate , k=22
L_khmer_5p = [] # 5% error rate, k=22
L_tallymer = [] # k=22, pick biggest memory, suffix 1 part and 4 parts different memory ,but all smaller than mkindex step. 
L_jellyfish_k22 = [] # use k=22, but memory change with different k size. k=31. 
L_jellyfish_k31 = []
L_dsk = [] # use k=22.

for i in range(1,6):

    L_khmer_1p.append(khmer_mem2[str(i)+'_1_22']/1000000) # memory usage of step 2 is always bigger than step 1
    L_khmer_5p.append(khmer_mem2[str(i)+'_5_22']/1000000) 
    L_tallymer.append(mkindex_mem[str(i)+'_1_22']/1000000) # memory usage of mkindex is always bigger than suffix(part1/4)
    if jelly_count_mem[str(i)+'_22'] > jelly_hist_mem[str(i)+'_22']:
        L_jellyfish_k22.append(jelly_count_mem[str(i)+'_22']/1000000)
    else:
        L_jellyfish_k22.append(jelly_hist_mem[str(i)+'_22']/1000000)

    if jelly_count_mem[str(i)+'_31'] > jelly_hist_mem[str(i)+'_31']:
        L_jellyfish_k31.append(jelly_count_mem[str(i)+'_31']/1000000)
    else:
        L_jellyfish_k31.append(jelly_hist_mem[str(i)+'_31']/1000000)
        
    L_dsk.append(dsk_mem[str(i)+'_22']/1000000)

#print L_khmer,L_tallymer,L_jellyfish_k22,L_jellyfish_k31,L_dsk

ax.set_xlabel('total number of distinct k-mers')
ax.set_ylabel('Memory Usage(G)')


ax.plot(total_list,L_khmer_1p,'bo-', label='khmer error rate=1%')
ax.plot(total_list,L_khmer_5p,'ko-', label='khmer error rate=5%')
ax.plot(total_list,L_tallymer,'go-', label='Tallymer')
ax.plot(total_list,L_jellyfish_k22,'ro-', label='Jellyfish k=22')
ax.plot(total_list,L_jellyfish_k31,'co-', label='Jellyfish k=31')
ax.plot(total_list,L_dsk,'yo-', label='DSK')

ax.set_xlim(0,2.5e9)
ax.set_ylim(0,40)
ax.legend(loc='best')
fig_file = '../figure/memory_benchmark.pdf'
plt.savefig(fig_file)
```

y-axis is the average of (offset/correct_count) for each k-mer

Since the high diversity dataset, most of the accurate counts are 1, so for smaller hash table(high error rate), the offset(miscount) may be 2 or 3. That percentage will be 100%-200%, which is the case in the figure while counting error rate >0.7

Figure 5
=======
Percentage of the unique k-mers starting in different position in reads

```python

file17 = open(datadir+"ecoli_ref.fastq.fasta.pos.abund=1",'r')
file32 = open(datadir+"ecoli_ref.fastq.fasta.pos32.abund=1",'r')

list1 = []
list2 = []
x1 = []
x2 = []

for line in file17:
    line = line.rstrip()
    fields = line.split()
    if fields[1] != '0':
        x1.append(float(fields[0]))
        list1.append(float(fields[1]))

for line in file32:
    line = line.rstrip()
    fields = line.split()
    if fields[1] != '0':
        x2.append(float(fields[0]))
        list2.append(float(fields[1]))

plt.xlabel("starting position of k-mer")
plt.ylabel("number of abund=1 k-mers at that position")
plt.title("low-abundance k-mer distribution within reads")


plt.plot(x1,list1,'r-')
plt.plot(x2,list2,'b-')
plt.legend(('k=17','k=32'),)
plt.ylim(0,10000000)
plt.xlim(0,100)

plt.savefig("../figure/perc_unique_pos.pdf")
```
